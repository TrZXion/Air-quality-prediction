```python
# Filter cities with weird names
weird_cities = df[df['city'].str.contains(r'[^\w\s,./-]', na=False)]

print("Cities with weird names:")
print(weird_cities['city'].unique())
```

```python
coordinates = (25.5941,85.1376)
results = rg.search(coordinates)
print(results)
```

```python
def fix_weird_city_names(row):
    if row['city'] in weird_cities['city'].unique():
        if pd.notnull(row['latitude']) and pd.notnull(row['longitude']):
            result = rg.search((row['latitude'], row['longitude']))
            city = result[0]['name']
            return city
        else:
            return row['city']
    else:
        return row['city']

df['city'] = df.apply(fix_weird_city_names, axis=1)
```

```python
columns_to_check = ['no2_concentration']

# Create box plots for each column
for column in columns_to_check:
    plt.figure(figsize=(8, 6))
    plt.boxplot(df[column])
    plt.title(f'Box plot of {column}')
    plt.xlabel('Column')
    plt.ylabel('Values')
    plt.show()
```

```python
columns_to_check = ['no2_concentration']

# Calculate number of rows with outliers for each column
for column in columns_to_check:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    # Define the outlier boundaries
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Identify outliers
    outliers = df[(df[column] > 200)]
    print(outliers)
    df = df[df['no2_concentration'] <= 200]
    # Reset index after removing rows
    df.reset_index(drop=True, inplace=True)
    
    # Print removed outlier rows
    print("Outlier rows removed:")
    print(outliers)
```

### Label Encoding on Categorical Columns

```python
from sklearn.preprocessing import LabelEncoder

# Define columns to encode
categorical_columns = ['city', 'country_name','who_region', 'iso3']

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Encode categorical columns
for column in categorical_columns:
    df_copy = df.copy()
    df_copy[column + '_encoded'] = label_encoder.fit_transform(df[column])
    df = df_copy

# Display the encoded DataFrame
print(df.head())
```

```python
corr_matrix = df.corr()

lower_threshold = 0.40
upper_threshold = 0.55
highly_correlated_pairs = []


for i in range(len(corr_matrix.columns)):
    for j in range(i + 1, len(corr_matrix.columns)):
        correlation_coefficient = abs(corr_matrix.iloc[i, j])
        if lower_threshold <= correlation_coefficient <= upper_threshold:
            highly_correlated_pairs.append((corr_matrix.columns[i], corr_matrix.columns[j], correlation_coefficient))

# Display highly correlated column pairs
print("Highly correlated column pairs within the range {} to {}:".format(lower_threshold, upper_threshold))
for pair in highly_correlated_pairs:
    print(pair)
```

```python
# Set the index of the DataFrame to the 'year' column
encoded_df.set_index('year', inplace=True)

# Define the time window for rolling statistics (e.g., 3 years)
window_size = 3

# Compute rolling statistics for each air quality parameter
df_rolling_avg = encoded_df.rolling(window=window_size).mean()
df_rolling_min = encoded_df.rolling(window=window_size).min()
df_rolling_max = encoded_df.rolling(window=window_size).max()
df_rolling_std = encoded_df.rolling(window=window_size).std()

# Concatenate the rolling statistics into a single DataFrame
rolling_statistics = pd.concat([df_rolling_avg, df_rolling_min, df_rolling_max, df_rolling_std],
                                keys=['rolling_avg', 'rolling_min', 'rolling_max', 'rolling_std'], axis=1)

# Reset the index of the DataFrame
rolling_statistics.reset_index(inplace=True)

# Display the DataFrame with rolling statistics
print(rolling_statistics.head())
```
