```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read CSV file with specified encoding
df = pd.read_csv('my_data.csv', encoding='Windows-1252')
```

```python
# Display DataFrame
df.head()
```

```python
df.tail()
```

```python
# Data Exploration
print("Number of rows and columns:", df.shape)
print("\nColumn names and data types:\n", df.dtypes)
print("\nMissing values:\n", df.isnull().sum())
# Summary Statistics
print("\nSummary statistics for numerical columns:\n", df.describe())
```

```python
# Split 'Coordinates' column into separate 'Latitude' and 'Longitude' columns
df[['Latitude', 'Longitude']] = df['Coordinates'].str.split(',', expand=True)

# Convert 'Latitude' and 'Longitude' columns to numeric data type
df['Latitude'] = pd.to_numeric(df['Latitude'], errors='coerce')
df['Longitude'] = pd.to_numeric(df['Longitude'], errors='coerce')

# Impute missing values in 'Latitude' and 'Longitude' columns with mean values
mean_latitude = df['Latitude'].mean()
mean_longitude = df['Longitude'].mean()
df['Latitude'].fillna(mean_latitude, inplace=True)
df['Longitude'].fillna(mean_longitude, inplace=True)
```

```python
# Check if there are any missing values left
print(df.isnull().sum())
```

```python
# Define a function to fill missing values with the most frequent value in the group
def fill_missing_city(group):
    mode_value = group.mode()
    if not mode_value.empty:
        return group.fillna(mode_value.iloc[0])
    else:
        return group

# Group by 'Country Code' and 'Location', then fill missing 'City' values within each group
df['City'] = df.groupby(['Country Code', 'Location'])['City'].transform(fill_missing_city)

# If there are still missing values, fill them with the most frequent city name overall
df['City'].fillna(df['City'].mode()[0], inplace=True)
```

```python
# Check if there are any missing values left
print(df.isnull().sum())
```

```python
# Find rows where 'Country Label' is null
null_country_label = df[df['Country Label'].isnull()]

# Display the rows where 'Country Label' is null
print(null_country_label)
```

```python
# Get the top N cities by frequency
top_cities = df['City'].value_counts().nlargest(50)

# Plot the top N cities
plt.figure(figsize=(10, 6))
top_cities.plot(kind='bar')
plt.title('Top Cities by Frequency')
plt.xlabel('City')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```
